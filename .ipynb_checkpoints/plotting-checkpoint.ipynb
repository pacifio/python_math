```python
plt.plot([0, 2], [2,2], 'r')
plt.plot([0, 2], [0, 0], 'k')
plt.plot([0, 0], [0, 2], 'g')
plt.plot([2, 2], [0, 2], 'm')

plt.axis('square')
plt.axis([-3,5,-3,5])
plt.title('A colourful square')
plt.show()
```

```python
x = [-5, 5]
m = -.3
b = .1
y = m*np.array(x) + b

plt.plot(x, y, label="y=%sx+%s" % (m, b))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)
axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')
plt.legend()
plt.show()
```

```python
x = [-5, 5]
m = [.7, -1.25]
b = [-2, .75]

for i in range(len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x, y, label="y = %sx+%s" % (m[i], b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)
axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')
plt.legend()
plt.show()
```

```python
x = range(-3, 4)
y = np.zeros(len(x))
for i in range(len(x)):
    y[i] = 2-x[i]**2

plt.plot(x, y, 's-')
plt.show()
```

```python
x = np.linspace(-3, 50, 30)
y = 2+np.sqrt(abs(x))
plt.plot(x, y, 'm-')
plt.show()
```

```python
x = np.linspace(-4, 4, 100)
e = range(-1, 4)
for i in e:
    y = x**i
    plt.plot(x, y, label="$y=x^%s$" % (i), linewidth=4)
plt.legend()
plt.xlim(x[0], x[-1])
plt.ylim([-20, 20])
plt.show()
```

```python
from sympy.abc import x
y = x**2
p = sym.plotting.plot(y, show=False)
p.xlim = [0, 50]
p[0].line_color = "m"
p.title = "This is a nice plot"
p.show()
```

```python
from sympy.plotting import plot as symplot
x, a = sym.symbols('x,a')
expr = a/x
n = 1
p = symplot(expr.subs(a, n), (x, -5, 5), show=False)

p2 = symplot(expr.subs(a, n+1), show=False)
p.extend(p2)

p.ylim = [-5, 5]
p.show()
```

```python
x, a = sym.symbols('x,a')
y = a/(x**2-a)
p = symplot(y.subs(a, 1), (x, -5, 5), show=False)
for i in range(2, 5):
    p.extend(symplot(y.subs(a, i), (x, -5, 5), show=False))

p.ylim = [-10, 10]
p.show()
```

```python
# Making images using matrixes

A = [
    [1,2],
    [1,4]
]

plt.imshow(A)
plt.xticks([0, 1])
plt.yticks([.85, 1.04])
plt.show()
```

```python
A = np.zeros([10, 14], dtype=int)
for i in range(np.shape(A)[0]):
    for j in range(np.shape(A)[1]):
        A[i][j] = 3*i-4*j
        plt.text(j,i, A[i, j], horizontalalignment='center', verticalalignment='center')
        
# plt.set_cmap('Purples')
plt.imshow(A)
plt.plot([0, 3], [8, 2], 'r', linewidth=4)
plt.show()
```

```python
# Implementing a checkerboard in python
N = 10
A = np.zeros([N, N], dtype=int)
for i in range(np.shape(A)[0]):
    for j in range(np.shape(A)[1]):
        A[i][j] = (-1)**(i+j)
plt.imshow(A)
plt.set_cmap('gray')
plt.tick_params(labelleft=False)
plt.show()
```

```python
# Drawing more complex shapes

from matplotlib.patches import Polygon

y1 = np.array([[1,1],[2,3],[3,1]])
p1 = Polygon(y1,facecolor='m',alpha=0.3)

y2 = np.array([[2,2], [2.5, 4], [3.5, 1]])
p2 = Polygon(y2,facecolor='b', alpha=0.3)

fig, ax = plt.subplots()
ax.add_patch(p1)
ax.add_patch(p2)

ax.set_ylim([0,4])
ax.set_xlim([0, 4])

plt.show()
```

```python
x = np.linspace(-2, 2, 101)
f = -x**2

pnts = np.vstack((x,f)).T
p = Polygon(pnts,facecolor='g')

y = np.array([[-.5, -4], [-.5, -2.5], [.5, -2.5], [.5, -4]])

p1 = Polygon(y, facecolor='k')

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
# ax.set_ylim([-2, 2])
# ax.set_xlim([-2, 2])
# plt.show()

plt.plot(x, f, 'k')
plt.plot([-2, 2], [-4, -4])
plt.axis('off')
plt.show()
```
